```python
def gradient_descent(x0, y0, f, grad_f, alpha, num_iterations):
    """
    Parameters:
    x0, y0: Initial point for the descent.
    f: a function of two variables.
    grad_f: the gradient of f as a function that returns a vector.
    alpha: Learning rate.
    num_iterations: Number of iterations to perform.

    Returns:
    (x, y): The coordinates of the final point after gradient descent.
    """
    x, y = x0, y0  # Initialize x and y with the initial point
    for i in range(num_iterations):
        # Obtain the gradient of f at (x, y)
        grad_x, grad_y = grad_f(x, y)

        # Update x and y by taking a step in the opposite direction of the gradient
        x = x - alpha * grad_x
        y = y - alpha * grad_y

        return x, y

def fun_1(x, y):
    return x**2 + y**2

def grad_f_1(x, y):
    grad_x = 2 * x
    grad_y = 2 * y
    return grad_x, grad_y

# Test case 1: (x0, y0) = (0.1, 0.1),  = 0.1, max iterations = 10
x1, y1 = gradient_descent(0.1, 0.1, fun_1, grad_f_1, 0.1, 10)
print(f"Final point after gradient descent: ({x1}, {y1})")

# Test case 2: (x0, y0) = (-1, 1),  = 0.01, max iterations = 100
x2, y2 = gradient_descent(-1, 1, fun_1, grad_f_1, 0.01, 100)
print(f"Final point after gradient descent: ({x2}, {y2})")

import numpy as np

def fun_2(x, y):
    return 1 - np.exp(-x**2 - (y - 2)**2) - 2 * np.exp(-x**2 - (y + 2)**2)

def grad_f_2(x, y):
    grad_x = 2 * x * (np.exp(-x**2 - (y - 2)**2) + 2 * np.exp(-x**2 - (y + 2)**2))
    grad_y = 2 * (y - 2) * np.exp(-x**2 - (y - 2)**2) + 4 * (y + 2) * np.exp(-x**2 - (y + 2)**2)
    return grad_x, grad_y

# Test case 1: (x0, y0) = (0, 1),  = 0.01, max iterations = 10000
x1, y1 = gradient_descent(0, 1, fun_2, grad_f_2, 0.01, 10000)
print(f"Final point after gradient descent: ({x1}, {y1})")

# Test case 2: (x0, y0) = (0, -1),  = 0.01, max iterations = 10000
x2, y2 = gradient_descent(0, -1, fun_2, grad_f_2, 0.01, 10000)
print(f"Final point after gradient descent: ({x2}, {y2})")

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(X, Y)
z = 1 - np.exp(-x**2 - (y - 2)**2) - 2 * np.exp(-x**2 - (y + 2)**2)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
ax.set_title('3D surface plot of f2(x, y)')
plt.show()

```
